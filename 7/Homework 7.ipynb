## 1. Use your own dataset (create a train and a test set) and build 2 models: Logistic Regression and Decision Tree (shallow). Compare the test results.

### Import caesarian dataset from UCI 

```python
import sklearn as sk 
import pandas as pd
```

```python
data = pd.read_csv('../data/caesarian.csv')
data.head()
```

### Split into train and test

```python
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(data.drop(columns = 'caesarian', axis = 1), data.caesarian, train_size = .66)
```

### Run logistic regression and create predictions

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_train)
pred_test = model.predict(X_test)
```

### Accuracy of logistic model against training data

```python
from sklearn.metrics import (
    accuracy_score, 
    classification_report, 
    confusion_matrix, auc, roc_curve
)
```

```python
accuracy_score(y_train, pred)
```

```python
confusion_matrix(y_train, pred)
```

```python
print(classification_report(y_train, pred))
```

### Accuracy of logistic model against testing data

```python
accuracy_score(y_test, pred_test)
```

```python
confusion_matrix(y_test, pred_test)
```

```python
print(classification_report(y_test, pred_test))
```

```python
print(classification_report(y_test, pred_test))
```

## Decision Tree Model

### Run Decision Tree model (depth 2) and create predicted data

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth= 2)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_train)
pred_test = model.predict(X_test)
```

### Accuracy of decision tree model (depth 2) against training data

```python
accuracy_score(y_train, pred)
```

```python
confusion_matrix(y_train, pred)
```

```python
print(classification_report(y_train, pred))
```

### Accuracy of decision tree model (depth 2) against training data

```python
confusion_matrix(y_test, pred_test)
```

```python
print(classification_report(y_test, pred_test))
```

```python
print(classification_report(y_test, pred_test))
```
